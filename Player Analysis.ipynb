```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
player_stats = pd.read_csv('players_stats.csv')
```

```python
df_clean = player_stats.drop(["Unnamed: 0", "player_link"], axis = 1)
```

```python
df_clean
```

```python
df_clean.set_index("player_nick", inplace=True)
```

```python
df_clean
```

```python
df_clean = df_clean.query('maps_played>999')
df_clean
```

```python
df_clean.describe()
```

```python
df_clean.info()
```

```python
hs_per = df_clean['hs_percentage']
hs_p = []

for i in hs_per:
    p = i.replace('%','')
    hs_p.append(float(p))
```

```python
df_age = df_clean['player_age']
age = np.array(list(map(np.int, df_age)))
age_list = list(age)
```

```python
impact = []
for i in df_clean['impact']:
    impact.append(float(i))
```

```python
df_x = pd.DataFrame({'country':df_clean['player_country'], 'hs_percent':hs_p, 'age':age, 'impact':impact, 'kills':df_clean['total_kills'], 
                     'kd_ratio':df_clean['kd_ratio']})
```

```python
df_x
```

```python
df_x.sort_values(by=['age'])
```

```python
df_x.sort_values(by=['kd_ratio'], ascending=False)
```

```python
y = df_x['age'][:28].sort_values(ascending=False)

ax = df_x.groupby('country')['age'].median().sort_values(ascending=False).plot.bar(figsize = (16,8), color ='lightgreen', width = 0.9)

for i, v in enumerate(y):
    ax.text(i-.33,v/y[i],y[i],fontsize=12,color="red")

plt.title('Median of age players per country', pad=15, fontsize=20)
plt.ylabel('Age', fontsize=14)
plt.xlabel('x', color='w')
plt.show()
```
